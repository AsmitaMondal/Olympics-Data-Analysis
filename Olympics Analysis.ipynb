```python
#10 countries with most wins 
not_n_medals=athletes_df[(athletes_df['Medal'].notnull())]
most=not_n_medals.Team.value_counts().reset_index(name='Medal')
most.head(10)
```

```python
#total females over the years
female_total=athletes_df[(athletes_df.Sex=='F')]
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=female_total,palette="coolwarm")
plt.title('Women Participation')
```

```python
#female participation trends in summer olympics
female_s=athletes_df[(athletes_df.Season=='Summer') & (athletes_df.Sex=='F')][['Sex','Year']]
part=female_s.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:'F'].plot()
plt.title('Trend in Female Prticipation in Summer Olympics')
```

```python
#female participation trends in winter olympics
female_w=athletes_df[(athletes_df.Season=='Winter') & (athletes_df.Sex=='F')][['Sex','Year']]
part=female_w.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:'F'].plot()
plt.title('Trend in Female Prticipation in Winter Olympics')
```

```python
#total males over the years
male_total=athletes_df[(athletes_df.Sex=='M')]
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=male_total,palette="Spectral")
plt.title('Men Participation')
```

```python
#male participation trends in summer olympics
male_s=athletes_df[(athletes_df.Season=='Summer') & (athletes_df.Sex=='M')][['Sex','Year']]
part=male_s.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:'F'].plot()
plt.title('Trend in Male Prticipation in Summer Olympics')
```

```python
#male participation trends in winter olympics
male_w=athletes_df[(athletes_df.Season=='Winter') & (athletes_df.Sex=='M')][['Sex','Year']]
part=male_w.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:'F'].plot()
plt.title('Trend in Male Prticipation in Winter Olympics')
```
